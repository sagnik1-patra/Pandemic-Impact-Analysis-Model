```python
import os
import json
import joblib
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam

# ============================================
# PATHS
# ============================================

DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Pandemic Impact Analysis Model\IC_2.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Pandemic Impact Analysis Model"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ============================================
# LOAD DATA
# ============================================

df = pd.read_csv(DATA_PATH)

df_melted = df.melt(
    id_vars=["SN", "State", "District"],
    value_vars=["2017","2018","2019","2020","2021","2022"],
    var_name="Year",
    value_name="Value"
)

df_melted["Year"] = df_melted["Year"].astype(int)
yearly = df_melted.groupby("Year")["Value"].mean().reset_index()

# ============================================
# TRAIN TEST SPLIT
# ============================================

train = yearly[yearly["Year"] <= 2020]
test = yearly[yearly["Year"] > 2020]

# ============================================
# SCALE DATA
# ============================================

scaler = MinMaxScaler()
scaled = scaler.fit_transform(yearly[["Value"]])

X_data = []
y_data = []

for i in range(len(scaled)-1):
    X_data.append(scaled[i])
    y_data.append(scaled[i+1])

X_data = np.array(X_data)
y_data = np.array(y_data)

X_data = X_data.reshape((X_data.shape[0],1,1))

X_train = X_data[:-2]
y_train = y_data[:-2]

X_test = X_data[-2:]
y_test = scaler.inverse_transform(y_data[-2:].reshape(-1,1))

# =====================================================
# FITNESS FUNCTION
# =====================================================

def fitness_function(units, lr, epochs):

    model = Sequential([
        LSTM(units=int(units), activation='relu', input_shape=(1,1)),
        Dense(1)
    ])

    model.compile(optimizer=Adam(learning_rate=lr), loss='mse')
    model.fit(X_train, y_train, epochs=int(epochs), verbose=0)

    pred = model.predict(X_test)
    pred = scaler.inverse_transform(pred)

    rmse = np.sqrt(mean_squared_error(y_test, pred))
    return rmse

# =====================================================
# 1 AIS INITIAL POPULATION
# =====================================================

population_size = 5
population = []

for _ in range(population_size):
    individual = [
        random.randint(20,100),      # units
        random.uniform(0.0005,0.01), # learning rate
        random.randint(20,80)        # epochs
    ]
    population.append(individual)

# Evaluate initial population
fitness_scores = [fitness_function(*ind) for ind in population]

# =====================================================
# 2 CSA REFINEMENT
# =====================================================

for iteration in range(5):

    for i in range(population_size):

        # Crow memory update
        new_solution = [
            population[i][0] + random.randint(-10,10),
            population[i][1] + random.uniform(-0.001,0.001),
            population[i][2] + random.randint(-10,10)
        ]

        # Keep values in valid range
        new_solution[0] = max(10, min(150, new_solution[0]))
        new_solution[1] = max(0.0001, min(0.02, new_solution[1]))
        new_solution[2] = max(10, min(100, new_solution[2]))

        new_fitness = fitness_function(*new_solution)

        if new_fitness < fitness_scores[i]:
            population[i] = new_solution
            fitness_scores[i] = new_fitness

# =====================================================
# BEST SOLUTION
# =====================================================

best_index = np.argmin(fitness_scores)
best_solution = population[best_index]

best_units, best_lr, best_epochs = best_solution

print("Best Hyperparameters:", best_solution)

# =====================================================
# FINAL TRAIN WITH BEST PARAMS
# =====================================================

model = Sequential([
    LSTM(int(best_units), activation='relu', input_shape=(1,1)),
    Dense(1)
])

model.compile(optimizer=Adam(learning_rate=best_lr), loss='mse')
model.fit(X_train, y_train, epochs=int(best_epochs), verbose=0)

pred = model.predict(X_test)
pred_rescaled = scaler.inverse_transform(pred)

rmse = np.sqrt(mean_squared_error(y_test, pred_rescaled))
r2 = r2_score(y_test, pred_rescaled)

# =====================================================
# SAVE MODEL
# =====================================================

model.save(os.path.join(OUTPUT_DIR, "hybrid_pandemic_model.keras"))

# =====================================================
# SAVE RESULTS CSV
# =====================================================

results_df = pd.DataFrame({
    "Best_Units":[best_units],
    "Best_Learning_Rate":[best_lr],
    "Best_Epochs":[best_epochs],
    "RMSE":[rmse],
    "R2":[r2]
})

results_df.to_csv(os.path.join(OUTPUT_DIR, "hybrid_results.csv"), index=False)

# =====================================================
# SAVE PREDICTIONS JSON
# =====================================================

prediction_dict = {
    "Actual": y_test.flatten().tolist(),
    "Hybrid_Predicted": pred_rescaled.flatten().tolist()
}

with open(os.path.join(OUTPUT_DIR, "hybrid_predictions.json"), "w") as f:
    json.dump(prediction_dict, f, indent=4)

# =====================================================
# GRAPHS
# =====================================================

# Accuracy Graph
plt.figure()
plt.bar(["Hybrid AIS-CSA"], [rmse])
plt.title("Hybrid Model RMSE")
plt.savefig(os.path.join(OUTPUT_DIR, "hybrid_accuracy_graph.png"))
plt.show()
plt.close()

# Prediction Graph
plt.figure()
plt.plot(test["Year"], y_test, label="Actual")
plt.plot(test["Year"], pred_rescaled, label="Hybrid Predicted")
plt.legend()
plt.title("Hybrid Prediction Comparison")
plt.savefig(os.path.join(OUTPUT_DIR, "hybrid_prediction_graph.png"))
plt.show()
plt.close()

# Heatmap
corr = df_melted.pivot_table(
    index="District",
    columns="Year",
    values="Value"
).corr()

plt.figure()
sns.heatmap(corr, annot=True)
plt.title("Hybrid Year Correlation Heatmap")
plt.savefig(os.path.join(OUTPUT_DIR, "hybrid_heatmap.png"))
plt.show()
plt.close()

print(" Hybrid AIS + CSA Model Completed Successfully!")
```
