```python
import os
import json
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam

# ============================================
# PATH CONFIG
# ============================================

DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Pandemic Impact Analysis Model\IC_2.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Pandemic Impact Analysis Model"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ============================================
# LOAD DATA
# ============================================

df = pd.read_csv(DATA_PATH)

df_melted = df.melt(
    id_vars=["SN", "State", "District"],
    value_vars=["2017","2018","2019","2020","2021","2022"],
    var_name="Year",
    value_name="Value"
)

df_melted["Year"] = df_melted["Year"].astype(int)

# Aggregate yearly mean
yearly = df_melted.groupby("Year")["Value"].mean().reset_index()

# ============================================
# SCALE DATA
# ============================================

scaler = MinMaxScaler()
scaled = scaler.fit_transform(yearly[["Value"]])

X_data = []
y_data = []

for i in range(len(scaled)-1):
    X_data.append(scaled[i])
    y_data.append(scaled[i+1])

X_data = np.array(X_data)
y_data = np.array(y_data)
X_data = X_data.reshape((X_data.shape[0],1,1))

# Train = 2017-2020
# Test = 2021-2022
X_train = X_data[:-2]
y_train = y_data[:-2]
X_test = X_data[-2:]
y_test_scaled = y_data[-2:]

# Proper test years for plotting
test_years = yearly["Year"].values[-2:]

y_test = scaler.inverse_transform(y_test_scaled.reshape(-1,1))

# =====================================================
# FITNESS FUNCTION
# =====================================================

def fitness(position):

    units = int(position[0])
    lr = position[1]
    epochs = int(position[2])

    model = Sequential([
        LSTM(units, activation='relu', input_shape=(1,1)),
        Dense(1)
    ])

    model.compile(optimizer=Adam(learning_rate=lr), loss='mse')
    model.fit(X_train, y_train, epochs=epochs, verbose=0)

    pred = model.predict(X_test, verbose=0)
    pred = scaler.inverse_transform(pred)

    rmse = np.sqrt(mean_squared_error(y_test, pred))
    return rmse

# =====================================================
# 1 AIS INITIAL POPULATION
# =====================================================

pop_size = 5
population = []

for _ in range(pop_size):
    individual = [
        random.randint(20,120),       # LSTM units
        random.uniform(0.0005,0.01),  # learning rate
        random.randint(20,80)         # epochs
    ]
    population.append(individual)

population = np.array(population, dtype=float)

# =====================================================
# 2 PSO OPTIMIZATION
# =====================================================

velocity = np.zeros_like(population)

pbest = population.copy()
pbest_scores = np.array([fitness(ind) for ind in population])

gbest = pbest[np.argmin(pbest_scores)]
gbest_score = np.min(pbest_scores)

w = 0.7
c1 = 1.5
c2 = 1.5

convergence_curve = []

for iteration in range(5):

    for i in range(pop_size):

        r1 = random.random()
        r2 = random.random()

        velocity[i] = (
            w * velocity[i]
            + c1 * r1 * (pbest[i] - population[i])
            + c2 * r2 * (gbest - population[i])
        )

        population[i] += velocity[i]

        # Boundaries
        population[i][0] = np.clip(population[i][0], 10, 150)
        population[i][1] = np.clip(population[i][1], 0.0001, 0.02)
        population[i][2] = np.clip(population[i][2], 10, 100)

        score = fitness(population[i])

        if score < pbest_scores[i]:
            pbest[i] = population[i]
            pbest_scores[i] = score

    gbest = pbest[np.argmin(pbest_scores)]
    gbest_score = np.min(pbest_scores)

    convergence_curve.append(gbest_score)

# =====================================================
# FINAL TRAINING
# =====================================================

best_units = int(gbest[0])
best_lr = gbest[1]
best_epochs = int(gbest[2])

print("Best Hyperparameters:", gbest)

model = Sequential([
    LSTM(best_units, activation='relu', input_shape=(1,1)),
    Dense(1)
])

model.compile(optimizer=Adam(learning_rate=best_lr), loss='mse')
model.fit(X_train, y_train, epochs=best_epochs, verbose=0)

pred_scaled = model.predict(X_test, verbose=0)
pred_rescaled = scaler.inverse_transform(pred_scaled)

rmse = np.sqrt(mean_squared_error(y_test, pred_rescaled))
r2 = r2_score(y_test, pred_rescaled)

# =====================================================
# SAVE MODEL
# =====================================================

model.save(os.path.join(OUTPUT_DIR, "pis_pandemic_model.keras"))

# =====================================================
# SAVE RESULTS CSV
# =====================================================

results_df = pd.DataFrame({
    "Best_Units":[best_units],
    "Best_Learning_Rate":[best_lr],
    "Best_Epochs":[best_epochs],
    "RMSE":[rmse],
    "R2":[r2]
})

results_df.to_csv(os.path.join(OUTPUT_DIR, "pis_results.csv"), index=False)

# =====================================================
# SAVE PREDICTIONS JSON
# =====================================================

prediction_dict = {
    "Years": test_years.tolist(),
    "Actual": y_test.flatten().tolist(),
    "Predicted": pred_rescaled.flatten().tolist()
}

with open(os.path.join(OUTPUT_DIR, "pis_predictions.json"), "w") as f:
    json.dump(prediction_dict, f, indent=4)

# =====================================================
# GRAPHS (SHOW + SAVE)
# =====================================================

# Accuracy Graph
plt.figure()
plt.bar(["Hybrid AIS-PSO"], [rmse])
plt.title("Hybrid AIS-PSO RMSE")
plt.savefig(os.path.join(OUTPUT_DIR, "pis_accuracy_graph.png"))
plt.show()
plt.close()

# Prediction Graph (FIXED)
plt.figure()
plt.plot(test_years, y_test, marker='o', label="Actual")
plt.plot(test_years, pred_rescaled, marker='o', label="Predicted")
plt.legend()
plt.title("Hybrid AIS-PSO Prediction")
plt.savefig(os.path.join(OUTPUT_DIR, "pis_prediction_graph.png"))
plt.show()
plt.close()

# Heatmap
corr = df_melted.pivot_table(
    index="District",
    columns="Year",
    values="Value"
).corr()

plt.figure()
sns.heatmap(corr, annot=True)
plt.title("Hybrid AIS-PSO Heatmap")
plt.savefig(os.path.join(OUTPUT_DIR, "pis_heatmap.png"))
plt.show()
plt.close()

# Convergence Curve
plt.figure()
plt.plot(convergence_curve)
plt.title("PSO Convergence Curve")
plt.savefig(os.path.join(OUTPUT_DIR, "pis_convergence.png"))
plt.show()
plt.close()

print(" Hybrid AIS + PSO Model Completed Successfully!")
```
