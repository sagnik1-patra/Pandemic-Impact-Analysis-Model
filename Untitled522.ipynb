```python
import os
import json
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam

# ============================================
# PATH CONFIG
# ============================================

DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Pandemic Impact Analysis Model\IC_2.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Pandemic Impact Analysis Model"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ============================================
# LOAD DATA
# ============================================

df = pd.read_csv(DATA_PATH)

df_melted = df.melt(
    id_vars=["SN", "State", "District"],
    value_vars=["2017","2018","2019","2020","2021","2022"],
    var_name="Year",
    value_name="Value"
)

df_melted["Year"] = df_melted["Year"].astype(int)

# Aggregate yearly mean
yearly = df_melted.groupby("Year")["Value"].mean().reset_index()

# ============================================
# SCALE DATA
# ============================================

scaler = MinMaxScaler()
scaled = scaler.fit_transform(yearly[["Value"]])

X_data = []
y_data = []

for i in range(len(scaled)-1):
    X_data.append(scaled[i])
    y_data.append(scaled[i+1])

X_data = np.array(X_data)
y_data = np.array(y_data)
X_data = X_data.reshape((X_data.shape[0],1,1))

# Train: 20172020
# Test: 20212022
X_train = X_data[:-2]
y_train = y_data[:-2]
X_test = X_data[-2:]
y_test_scaled = y_data[-2:]
y_test = scaler.inverse_transform(y_test_scaled.reshape(-1,1))

test_years = yearly["Year"].values[-2:]

# =====================================================
# FITNESS FUNCTION
# =====================================================

def fitness(position):

    units = int(position[0])
    lr = position[1]
    epochs = int(position[2])

    model = Sequential([
        LSTM(units, activation='relu', input_shape=(1,1)),
        Dense(1)
    ])

    model.compile(optimizer=Adam(learning_rate=lr), loss='mse')
    model.fit(X_train, y_train, epochs=epochs, verbose=0)

    pred = model.predict(X_test, verbose=0)
    pred = scaler.inverse_transform(pred)

    rmse = np.sqrt(mean_squared_error(y_test, pred))
    return rmse

# =====================================================
# 1 CSA INITIALIZATION (Exploration)
# =====================================================

pop_size = 6
population = []

for _ in range(pop_size):
    individual = [
        random.randint(20,120),        # LSTM units
        random.uniform(0.0005,0.01),   # learning rate
        random.randint(20,80)          # epochs
    ]
    population.append(individual)

population = np.array(population, dtype=float)
fitness_scores = np.array([fitness(ind) for ind in population])

best_index = np.argmin(fitness_scores)
best = population[best_index].copy()
best_score = fitness_scores[best_index]

# =====================================================
# 2 BAT ALGORITHM REFINEMENT (Exploitation)
# =====================================================

velocity = np.zeros_like(population)
frequency = np.zeros(pop_size)

convergence_curve = []

for iteration in range(6):

    for i in range(pop_size):

        frequency[i] = random.uniform(0,2)

        velocity[i] += (population[i] - best) * frequency[i]
        new_solution = population[i] + velocity[i]

        # Boundaries
        new_solution[0] = np.clip(new_solution[0], 10,150)
        new_solution[1] = np.clip(new_solution[1], 0.0001,0.02)
        new_solution[2] = np.clip(new_solution[2], 10,100)

        new_score = fitness(new_solution)

        if new_score < fitness_scores[i]:
            population[i] = new_solution
            fitness_scores[i] = new_score

        if new_score < best_score:
            best = new_solution.copy()
            best_score = new_score

    convergence_curve.append(best_score)

# =====================================================
# FINAL TRAINING
# =====================================================

best_units = int(best[0])
best_lr = best[1]
best_epochs = int(best[2])

print("Best Hyperparameters:", best)

model = Sequential([
    LSTM(best_units, activation='relu', input_shape=(1,1)),
    Dense(1)
])

model.compile(optimizer=Adam(learning_rate=best_lr), loss='mse')
model.fit(X_train, y_train, epochs=best_epochs, verbose=0)

pred_scaled = model.predict(X_test, verbose=0)
pred_rescaled = scaler.inverse_transform(pred_scaled)

rmse = np.sqrt(mean_squared_error(y_test, pred_rescaled))
r2 = r2_score(y_test, pred_rescaled)

# =====================================================
# SAVE MODEL
# =====================================================

model.save(os.path.join(OUTPUT_DIR, "bsa_pandemic_model.keras"))

# =====================================================
# SAVE RESULTS CSV
# =====================================================

results_df = pd.DataFrame({
    "Best_Units":[best_units],
    "Best_Learning_Rate":[best_lr],
    "Best_Epochs":[best_epochs],
    "RMSE":[rmse],
    "R2":[r2]
})

results_df.to_csv(os.path.join(OUTPUT_DIR, "bsa_results.csv"), index=False)

# =====================================================
# SAVE PREDICTIONS JSON
# =====================================================

prediction_dict = {
    "Years": test_years.tolist(),
    "Actual": y_test.flatten().tolist(),
    "Predicted": pred_rescaled.flatten().tolist()
}

with open(os.path.join(OUTPUT_DIR, "bsa_predictions.json"), "w") as f:
    json.dump(prediction_dict, f, indent=4)

# =====================================================
# GRAPHS (SHOW + SAVE)
# =====================================================

# Accuracy Graph
plt.figure()
plt.bar(["Hybrid CSA-BA"], [rmse])
plt.title("Hybrid CSA-BA RMSE")
plt.savefig(os.path.join(OUTPUT_DIR, "bsa_accuracy_graph.png"))
plt.show()
plt.close()

# Prediction Graph
plt.figure()
plt.plot(test_years, y_test, marker='o', label="Actual")
plt.plot(test_years, pred_rescaled, marker='o', label="Predicted")
plt.legend()
plt.title("Hybrid CSA-BA Prediction")
plt.savefig(os.path.join(OUTPUT_DIR, "bsa_prediction_graph.png"))
plt.show()
plt.close()

# Heatmap
corr = df_melted.pivot_table(
    index="District",
    columns="Year",
    values="Value"
).corr()

plt.figure()
sns.heatmap(corr, annot=True)
plt.title("Hybrid CSA-BA Heatmap")
plt.savefig(os.path.join(OUTPUT_DIR, "bsa_heatmap.png"))
plt.show()
plt.close()

# Convergence Curve
plt.figure()
plt.plot(convergence_curve)
plt.title("BA Convergence Curve")
plt.savefig(os.path.join(OUTPUT_DIR, "bsa_convergence.png"))
plt.show()
plt.close()

print(" Hybrid CSA + BA Model Completed Successfully!")
```
